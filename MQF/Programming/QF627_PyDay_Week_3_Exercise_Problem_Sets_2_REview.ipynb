```python
def visualize_default_rate_by_cat(df,
                                  cat,
                                  ax = None,
                                  set_title = True):
    if set_title:
        title = f"Default Rate per {cat}"
    else:
        title = None

    (
        df
        .groupby(by = cat) # "grade" as parameter to input
        .loan_status
        .sum()
        /
        df
        .groupby(by = cat)
        .loan_status
        .count()
        * 100
        ).plot\
         .bar(title = title, # defined here with you string input
              ylabel = "Default Rate (unit: %)",
              ax = ax)
    
    if ax == None:
        plt.show()
```

#### Question 5.3 Write a function which takes in a categorical variable and computed the average default rate across the categories.

> You can also compute the `difference between the highest and the lowest default rate` across the categories, which is a decent metric indicating the effect of the varaible on default rate.

> Please create your own function named `diff_rate`.

```python
def obtain_diff_rate(df,
                     category): # you will analyze by category
    df_SUBSET =\
    (
        df
        .groupby(by = category)
        .loan_status
        .sum()
        /
        df
        .groupby(by = category)
        .loan_status
        .count()
    )

    diff_rate =\
        (
            df_SUBSET.max()
            -
            df_SUBSET.min()
        )

    return diff_rate, df_SUBSET.idxmax(), df_SUBSET.idxmin()
```
