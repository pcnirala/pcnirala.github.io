### Making new data columns
> You can make a new column based on two columns like so:

```python
tabular_data.columns
```

> To overwrite your previous data you can just re-assign the column new values, the way you do with any variable: 

```python
# tabular_data["asian_prop"] = tabular_data["asian"] / tabular_data.total_population * 100
```

```python
import this
```

```python
tabular_data["asian_prop"] =\
( # flow programming (chaining)
    tabular_data.asian
    /
    tabular_data.total_population
    * 100
)
```

##### A quick sorting function can now help you find the spots with the highest or lowest black populations (we can go over this again next week):

```python
(
    tabular_data
    [["asian_prop", "state"]]
    .sort_values(by = "asian_prop") # team, by default, it's ascending order.
)
```

```python
(
    tabular_data
    [["asian_prop", "state"]]
    .sort_values(by = "asian_prop",
                 ascending = False)
)
```

### Grouping your data results

> `.groupby()`

```python
tabular_data
```

```python
(
    tabular_data
    .asian_prop
    .mean()
)
```

```python
(
    tabular_data
    .groupby(["state"]) # re-set unit of analysis at state-level
    .asian_prop
    .mean()
).to_frame()
```

```python
(
    tabular_data
    .groupby(["county"]) # re-set unit of analysis at `county`-level
    .asian_prop
    .mean()
)
```

> `.agg()`

    recall that between dict. and list, dict. is more pythonian data type!

```python
(
    tabular_data
    .groupby(["state"])
    .agg({"asian_prop": "mean"}
        )
)
```
