### Please import necessary packages (modules, libraries; i.e., `numpy`, `matplotlib`, and `pandas`)

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 

%matplotlib inline
```

### Question 1. Under the `Problem-sets for Exercise` tab you will find 12 CSV files. Please read the 12 CSV files. Then, please merge the 12 CSV files into a single dataset.

```python
month_list =\
[
    "January", 
    "February", 
    "March", 
    "April", 
    "May", 
    "June", 
    "July", 
    "August", 
    "September", 
    "October", 
    "November", 
    "December"
]

dataframes = []
for counter in range(12):
    df =\
    (
        pd
        .read_csv(
            "Exercise_Data_" 
            + month_list[counter] 
            + ".csv"
        )
    )
    dataframes.append(df)

merged_df =\
(
     pd
    .concat(
        dataframes
    )
)

merged_df
```

### <font color="green">***Now let's clean up the dataset :)***</font>

> The first step in this exercise is figuring out what we need to clean.

### Question 2. Please remove rows of NAN

```python
merged_df_nan_removed =\
(
    merged_df
    .dropna()
)
merged_df_nan_removed
```

### Question 3. Please remove text in `Order Date` column

```python
merged_df_copy =\
(
    merged_df_nan_removed
    .copy()
) 
merged_df_copy['Order Date'] =\
(
    pd
    .to_datetime(
        merged_df_copy['Order Date'], 
        format='%m/%d/%y %H:%M',
        errors='coerce'
    )
)
merged_df_order_date_cleaned =\
(
    merged_df_copy
    .dropna(
        subset=['Order Date']
    )
)
merged_df_order_date_cleaned
```

### Question 4. Please make columns correct type: namely, make `Quantity Ordered` and `Price Each` as numeric data types.

```python
merged_df_copy =\
(
    merged_df_order_date_cleaned
    .copy()
)
merged_df_copy['Quantity Ordered'] =\
    pd.to_numeric(
        merged_df_copy['Quantity Ordered'],
        errors='coerce'
    )

merged_df_copy['Price Each'] =\
    pd.to_numeric(
        merged_df_copy['Price Each'],
        errors='coerce'
    )

merged_df_cleaned =\
(
    merged_df_copy
    .dropna(
        subset=['Quantity Ordered',
                'Price Each']
    )
)

merged_df_cleaned
```

### <font color="green">***Let's create additional columns (variables) for your data exploration.</font>***

### Question 5. Please add `Month` column: Create the `Month` from `Order Date` column

```python
merged_df_with_month =\
(
    merged_df_cleaned
    .copy()
)
merged_df_with_month['Month'] =\
(
    merged_df_with_month[
        'Order Date'
    ]
    .dt
    .month
)
merged_df_with_month
```

### Question 6. Please add `City` column: Create the `City` from `Purchase Address` column

```python
merged_df_with_city =\
(
    merged_df_with_month
    .copy()
)
merged_df_with_city['City'] =\
(
    merged_df_with_city[
        'Purchase Address'
    ]
    .apply(
        lambda x: x.split(',')[1]
        .strip()
    )
)
merged_df_with_city
```

### <font color="green">***Now, let's explore our data***</font>

### Question 7: What was the best month for the business? How much was earned that month? Please use `Quantity Ordered` and `Price Each` columns to create a new column (variable) `Gross Revenue`.

```python
merged_df_with_gross_revenue =\
(
    merged_df_with_city
    .copy()
)
merged_df_with_gross_revenue['Gross Revenue'] =\
(
    merged_df_with_gross_revenue['Quantity Ordered'] 
    * 
    merged_df_with_gross_revenue['Price Each']
)

monthly_gross_revenue =\
(
    merged_df_with_gross_revenue
    .groupby(
        'Month'
    )
    [
        'Gross Revenue'
    ]
    .sum()  
) 

best_month =\
(
    monthly_gross_revenue
    .idxmax()
)

best_month_revenue =\
(
    monthly_gross_revenue[best_month]
)

print("Best Month: " + str(best_month) + ", Best Month Revenue: " + str(best_month_revenue))
```

### Question 8. Please create a vertical bar plot that summarizes the `monthly` `gross revenue`.

```python
plt.figure(figsize=(4, 2))
plt\
    .bar(
        monthly_gross_revenue.index, 
        monthly_gross_revenue.values, 
        color='green'
    )

plt.xlabel('Month')
plt.ylabel('Gross Revenue ($)')
plt.title('Monthly Gross Revenue')
plt.xticks(monthly_gross_revenue.index)

plt.show()
```

### Question 9: What `City` sold the most product based on `Quantity Ordered` and `Gross Revenue`?

```python
city_sales =\
(
    merged_df_with_gross_revenue
    .groupby(
        'City'
    )
    .agg({
        'Quantity Ordered': 'sum',
        'Gross Revenue': 'sum'
    })
    .reset_index()
)

city_with_most_quantity =\
(
    city_sales
    .loc[
        city_sales['Quantity Ordered']
        .idxmax()
    ]
)

city_with_most_revenue =\
(
    city_sales
    .loc[
        city_sales['Gross Revenue']
        .idxmax()
    ]
)

if city_with_most_quantity['City'] == city_with_most_revenue['City']:
    city_with_most_sales =\
        city_with_most_quantity['City']
else:
    #No single city satisfies both conditions
    city_with_most_sales = None

city_with_most_sales
```

### Question 10. Please visualize the above results with `horizontal` bar plot.

```python
city_sales_sorted_quantity =\
(
    city_sales
    .sort_values(
        'Quantity Ordered', 
        ascending=False
    )
)

city_sales_sorted_revenue =\
(
    city_sales
    .sort_values(
        'Gross Revenue',
        ascending=False
    )
)

# Quantity Ordered by City
fig1, ax1 =\
(
    plt
    .subplots(
        figsize=(4, 2)
    )
)
ax1\
    .barh(
        city_sales_sorted_quantity['City'],
        city_sales_sorted_quantity['Quantity Ordered'],
        color='green'
    )
ax1.set_xlabel('Quantity Ordered')
ax1.set_ylabel('City')
ax1.set_title('Total Quantity Ordered by City')

# Gross Revenue by City
fig2, ax2 =\
(
    plt
    .subplots(
        figsize=(4, 2)
    )
)
ax2\
    .barh(
        city_sales_sorted_revenue['City'],
        city_sales_sorted_revenue['Gross Revenue'],
        color='blue'
    )
ax2.set_xlabel('Gross Revenue ($)')
ax2.set_ylabel('City')
ax2.set_title('Total Gross Revenue by City')

city_with_most_sales = (fig1, fig2)
```

### Question 11: What time should we display ads to maximize likelihood of customer's buying product? Please create a new variable `Hour` from column `Order Date`

```python
merged_df_with_hour =\
(
    merged_df_with_gross_revenue
    .copy()
)

merged_df_with_hour['Hour'] =\
(
    merged_df_with_hour[
        'Order Date'
    ]
    .dt
    .hour
)

hourly_orders_count =\
(
    merged_df_with_hour
    .groupby('Hour')
    .size()  
) 

best_hour_for_ads =\
(
    hourly_orders_count
    .idxmax()
)

best_hour_for_ads
```

### Question 12. Please visualize the results of your analysis above.

```python
plt.figure(figsize=(10, 6))
plt\
    .bar(
        hourly_orders_count.index,
        hourly_orders_count.values
    )
plt.xlabel('Hour of the Day')
plt.ylabel('Number of Orders')
plt.title('Number of Orders by Hour')
plt.xticks(hourly_orders_count.index)
plt.show()
```
