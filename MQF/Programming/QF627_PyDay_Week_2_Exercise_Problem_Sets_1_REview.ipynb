### Question 8. Let's write a custom function that gives us a `dictionary` that lists `talks` containing `funny` rating.

````python
import this
````

````python
def make_dict_funny(df: pd.DataFrame):

    DF_talk_new_LIST =\
        (
            df
            .loc[ : , ["title", "new_LIST"]
                 ]
        )

    placeholder_dict = {}

    for A, B in DF_talk_new_LIST.iterrows():

        title = B["title"] # note that we use .iterrows() here :)

        dict_for_ratings = B["new_LIST"]

        for k in dict_for_ratings:

            if k["name"].lower() == "funny" and k["count"] > 0:

                placeholder_dict[title] = k

    return placeholder_dict
````

> Then, you might want to examine a record in which "Funny" is not the first dictionary

````python
DF_for_funny =\
(
    pd
    .DataFrame(make_dict_funny(ted)
              )
)
````
